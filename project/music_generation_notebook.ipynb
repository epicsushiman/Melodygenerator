# Melody Generator: Generative Music Notebook

This notebook generates three MIDI files and three WAV audio renders using Euclidean rhythms, Markov-style melodic movement, seeded randomness, and a small Python synthesizer.

```python
from pathlib import Path
from scripts.generate_variations import VARIATIONS, compose, write_midi, render_wav

output_root = Path('output')
midi_dir = output_root / 'midi'
audio_dir = output_root / 'audio'
midi_dir.mkdir(parents=True, exist_ok=True)
audio_dir.mkdir(parents=True, exist_ok=True)

[(v.title, v.tempo, v.seed, v.melody_pulses, v.melody_steps) for v in VARIATIONS]
```

## Algorithm Sketch

- Euclidean rhythm distributes a chosen number of hits across a fixed step grid.
- A weighted melodic transition chooses the next scale degree from nearby steps and occasional leaps.
- Seed values make each take repeatable while preserving variation between takes.
- The Python generator defaults to D Dorian, while the browser demo can shift key and mode live.
- MIDI is exported directly, then a simple sine/noise synth renders WAV audio.

```python
created_files = []
for spec in VARIATIONS:
    events = compose(spec)
    midi_path = midi_dir / f'{spec.slug}.mid'
    wav_path = audio_dir / f'{spec.slug}.wav'
    write_midi(events, spec.tempo, midi_path)
    render_wav(events, spec.tempo, wav_path)
    created_files.append((midi_path, wav_path, len(events)))

created_files
```

```python
# Optional playback in Jupyter/Colab after generation:
try:
    from IPython.display import Audio, display
    for _, wav_path, _ in created_files:
        print(wav_path)
        display(Audio(str(wav_path)))
except Exception as exc:
    print('Playback display is optional:', exc)
```
